```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('datasets/salary_data_cleaned.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
# Histograms

columns = ['Rating', 'Founded', 'hourly', 'salary', 'age', 'python_yn', 'aws_yn', 'excel_yn', 'communication_yn', 'desc_len']

fig, axs = plt.subplots(len(columns), figsize=(8, 40))
fig.subplots_adjust(hspace=0.3)
for n, col in enumerate(columns):
    sns.histplot(df, x=col, ax=axs[n])
```

```python
df[['age', 'salary', 'Rating']].corr()
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df[['age','salary','Rating','desc_len']].corr(),vmax=.3, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
df.columns
```

```python
df_cat = df[['Location', 'Size', 'Type of ownership',
       'Industry', 'Sector', 'Revenue', 'company_txt', 
             'python_yn', 'aws_yn', 'excel_yn', 'communication_yn', 'job_title', 'seniority']]
```

```python
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    plt.subplots(figsize=(20,5))
    print(f"Graph for {i}: total = {len(cat_num)}")
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```
